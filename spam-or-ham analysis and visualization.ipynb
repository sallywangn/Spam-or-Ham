```python
# import dependancies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import seaborn as sns
```

```python
#read the dataframe
df = pd.read_csv("emails.csv", sep = ",")
df
```

```python
# Drop the email column as it is not needed
df = df.drop(columns=['Email No.'])
```

```python
# for null column
null_counts = df.isnull().sum()
# Filter columns with more than one null value
columns_with_nulls = null_counts[null_counts >= 1]
columns_with_nulls
```

```python
# count of unique items
df.nunique()
```

```python
#drop duplicates then create another dataframe excluding duplicates
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
# Heatmap 
sns.heatmap(df.isnull())
```

```python
# Determining Statistics
for column in df:
    print(f"{column}:")
    print(f"  Min: {df[column].min()}")
    print(f"  Max: {df[column].max()}")
    print(f"  Mean: {df[column].mean()}")
    print(f"  Variance: {df[column].var()}")
    print(f"  Standard Deviation: {df[column].std()}")
```

```python
# Density plot for the first ten columns
for i in df.columns[:10] :
    sns.histplot(df[i])
    plt.title(i)
    plt.show()
```
